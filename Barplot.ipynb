```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
g = sns.FacetGrid(iris.melt(id_vars="species"), col="variable", col_wrap=2, sharex=False, sharey=False)
g.map(sns.barplot, "species", "value", "species", order=iris["species"].unique(), palette="Set2", ci=None)
```

```python
for ax, title in zip(g.axes.flatten(), ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]):
    ax.set_title(f"Average {title}")
    ax.set_xlabel("Species")
    ax.set_ylabel(f"{title} (cm)")
```

```python
g.fig.tight_layout()
plt.show()
```
